```python
import os, sys, csv, json, xlrd

#globals 

INPUT_FILE='module-master-feb-matt-feb22-updated-vikas-may29-2018.xlsx'
OUTPUT_FILE_NAME='all_module_config_test_may29.json'
CERTIFICATION_FILE_NAME='certification.conf'
CERTIFICATION_MODULE='Module-C'

def create_module_config_v2(module_data_stuff):
    
    #modules = set([rows[0] for rows in module_data[1:] ])
    units = set([str(rows[1]).split('.')[0] for rows in module_data_stuff[1:] ])
    #print units
    
    d={}
    
    for u in units:
        d[u] = {}
        pageseq = []
        d[u]['page_sequence'] = pageseq
        topics =[]
        d[u]['topics'] = topics

        for row in module_data_stuff[1:]:
            #print row
            if str(row[1]).split('.')[0] == u:
                d[u].setdefault('title', row[2])
                d[u]['page_sequence'].append(row[5])
                d[u]['topics'].append(row[3])
        d[u]['footer'] = zip(d[u]['page_sequence'], d[u]['topics'])
        
        start = '<span class="footerspan">Unit: </span><select id="footer_selector" onchange="jumpTo(); return false; " class="footer_selector" >'
        end = '</select>'
        for i in d[u]['footer']:
             start += '<option value="' + i[0] + '">' + i[1] + '</option>' 

        d[u]['footerNav'] = start + end

        #create jump to module menu 
        
        
        #create side menu bar . 
        
        

    #print d[3.0]['footerNav']
    
    return d

##############################################################################################
### MAIN METHOD ##########
#wkbook = xlrd.open_workbook('module-master-feb-matt-feb22-updated-vikas-may29-2018.xlsx')
wkbook = xlrd.open_workbook(INPUT_FILE)
module_ids = wkbook.sheet_names()
print module_ids

module_config = {}
module_config['mods'] = {}
module_config['meta'] = {}

#get module titles
mod_titles = wkbook.sheet_by_name('ModuleTitles')
mod_list = [mod_titles.row_values(r) for r in xrange(mod_titles.nrows)]
mod_titles= {}
#mod_names = defaultdict(lambda: 'My default')
for n in mod_list:
    mod_titles[n[0]] = n[1]
    
#MODULES TO IGNORE

ignore = ['Module-8', 'ModuleTitles' ]

"""
for w in range(len(wkbook.sheets())):
    print w
"""
for m in module_ids:
    print m 
    if m not in ignore:
        current_module = wkbook.sheet_by_name(m)
        print current_module.name
        if current_module:
            module_config['mods'][current_module.name] = {}
            current_module_data = [current_module.row_values(r) for r in xrange(current_module.nrows)]
            module_config['mods'][current_module.name]['conf'] = create_module_config_v2(current_module_data)
    
    #module_config.append(current_module_data)
    
    #print current_module_data
    
    
#Construct side menu. 

#get number of modules 
side = '<ul id="menu"><li><a href="home.html">Welcome</a></li>' 
#for r in range(len(module_config['mods'])):
for r in module_ids:
    #print r
    if r not in ignore:
        module_title = r # 'Module-'+ str(r)
        module_title_description = mod_titles[module_title]
        #side+= '<li><a href="' + module_title  + '.html">' + module_title + '</a> + <ul id="submenu">'
        #side+= '<li><a href="#">' + module_title + '</a> + <ul id="submenu">' 
        side+= '<li><a href="#" id="' + module_title + '"><span id="side-bullet" class="side-bullet" style="margin-right:5px;"></span>' + module_title_description  + '</a> + <ul id="submenu">' 
        side+='<li><a href="home.html?loadPage=' + module_title + '.html' + '">' + 'Module Home' + '</a><li>' 
        module = module_config['mods'][module_title] # ['Module-'+ str(r)]
        for u in range(len(module['conf'])):
            if str(u) in module['conf'] :
                unit = module['conf'][str(u)]
                title = unit['title']
                first_slide = unit['page_sequence'][0]
                mod_num = module_title.split('-')[-1]
                id = first_slide
                #id = mod_num + '_'+ str(u) + '_' + first_slide # 'slide01.html'
                side+= '<li><a href="#" id="' + id  + '"'+  ' onClick="load_page_menu(this.id)"' + ' ><span id="side-bullet" style="margin-right:5px;">&#9634;</span> ' + str(u) + ' : ' + title + '</a></li>'

        side+='</ul></li>'

side += '</ul>'

#print side

module_config['meta']['side_menu'] = side

print module_config['meta']['side_menu']

print module_config['mods']['Module-5']['conf']['3']['footerNav']
          
with open(OUTPUT_FILE_NAME, 'w') as f:
    f.write(json.dumps(module_config))
    
#create certification module 

certification_mod = module_config['mods'][CERTIFICATION_MODULE]
module_config.clear()
module_config['mods']= certification_mod

with open(CERTIFICATION_FILE_NAME, 'w') as f:
    f.write(json.dumps(module_config))
```
